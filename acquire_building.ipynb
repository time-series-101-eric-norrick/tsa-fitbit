### I noticed what seperates these two categories is a newline char, so I've gone ahead and made that the point at which we stop capturing for "food"

```python
def grab_foods(file):
    foods = []
    for line in file:
        if line == "":
            break
        foods.append(line)
    foods = foods[1:]
    return foods


def remove_foods(file):
    file = file.copy()
    while True:
        if file.pop(0) == '':
            break
    
    file.pop(0) ## Removes the "Activites"
    file.pop(0) ## Removes "Column names"
    
    return file  
    
```

```python
def grab_activities(file):
    file = file.copy()
    file = remove_foods(file)
    activities = {}
    for string in file:
        ### This checks to see if we hit the next empty string signifying we've oved on to food logs
        if string == "":
            break
        x = string.split('"')
        activities[x[1]] = {'total_calories':x[3],
                            'steps':x[5],
                           'distance':x[7],
                           'floors':x[9],
                           'mins_sedentary':x[11],
                           'mins_active_light':x[13],
                           'mins_active_med':x[15],
                           'mins_active_high':x[17],
                           'activity_calories':x[19]
                           } 
        
    return activities
```

```python
activities = grab_activities(file)
```

```python
def remove_activities(file):
    file = file.copy()
    file = remove_foods(file)
    while True:
        if file.pop(0) == '':
            break
            
    return file
```

```python
def grab_food_logs(file):
    file = file.copy()
    file = remove_foods(file)
    file = remove_activities(file)
    food_logs = {}
    while True:
        food_log = file.pop(0).split()[2]
        file.pop(0)
        calories = file.pop(0).split(',')[2].replace('"','')
        fat = file.pop(0).split(',')[2].replace('"','')
        fiber = file.pop(0).split(',')[2].replace('"','')
        carbs = file.pop(0).split(',')[2].replace('"','')
        sodium = file.pop(0).split(',')[2].replace('"','')
        protien = file.pop(0).split(',')[2].replace('"','')
        water= file.pop(0).split(',')[2].replace('"','')
        file.pop(0)
    
        food_logs[food_log] ={'calories' : calories,
                         'fat' : fat,
                         'fiber' : fiber,
                         'carbs': carbs,
                         'sodium' : sodium,
                         'protien' :protien,
                         'water' : water}
        if len(file) == 0:
            return food_logs
            
```

```python
food_logs = grab_food_logs(file)
```
